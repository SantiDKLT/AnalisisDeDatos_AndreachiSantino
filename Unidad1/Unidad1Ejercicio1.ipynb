```python
def EliminarEsp(lista):
  ListaS_Esp = []
  for elemento in lista:
    ElementoSEsp = elemento.replace(" ", "")
    ListaS_Esp.append(ElementoSEsp)
  return ListaS_Esp

lista = []
while True:
  elemento = input("Ingrese un elemento (o 'fin' para terminar): ")
  if elemento == "fin":
    break
  lista.append(elemento)

ListaS_Esp = EliminarEsp(lista)

print("Lista original:")
for elemento in lista:
  print(f"- {elemento}")

print("\nLista sin espacios:")
for elemento in ListaS_Esp:
  print(f"- {elemento}")
```
